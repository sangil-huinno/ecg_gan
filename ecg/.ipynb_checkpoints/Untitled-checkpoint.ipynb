```python
import pandas as pd
import os
from matplotlib import pyplot as plt
import numpy as np
from numpy import genfromtxt
```

```python
my_datal11 = genfromtxt('Generatornormall1_img/Generatornormall1_correler.txt', delimiter='')
my_datal12 = genfromtxt('Generatornormall1_img/Generatornormall1_dotter.txt', delimiter='')
my_datal13 = genfromtxt('Generatornormall1_img/Generatornormall1_dloss.txt', delimiter='')
my_datal14 = genfromtxt('Generatornormall1_img/Generatornormall1_gloss.txt', delimiter='')

my_datal21 = genfromtxt('Generatornormall2_img/Generatornormall2_correler.txt', delimiter='')
my_datal22 = genfromtxt('Generatornormall2_img/Generatornormall2_dotter.txt', delimiter='')
my_datal23 = genfromtxt('Generatornormall2_img/Generatornormall2_dloss.txt', delimiter='')
my_datal24 = genfromtxt('Generatornormall2_img/Generatornormall2_gloss.txt', delimiter='')

my_data1 = genfromtxt('Generatornormalbase_img/Generatornormalbase_correler.txt', delimiter='')
my_data2 = genfromtxt('Generatornormalbase_img/Generatornormalbase_dotter.txt', delimiter='')
my_data3 = genfromtxt('Generatornormalbase_img/Generatornormalbase_dloss.txt', delimiter='')
my_data4 = genfromtxt('Generatornormalbase_img/Generatornormalbase_gloss.txt', delimiter='')

my_datac1 = genfromtxt('Generatornormalcorrel0_img/Generatornormalcorrel0_correler.txt', delimiter='')
my_datac2 = genfromtxt('Generatornormalcorrel0_img/Generatornormalcorrel0_dotter.txt', delimiter='')
my_datac3 = genfromtxt('Generatornormalcorrel0_img/Generatornormalcorrel0_dloss.txt', delimiter='')
my_datac4 = genfromtxt('Generatornormalcorrel0_img/Generatornormalcorrel0_gloss.txt', delimiter='')

my_datad1 = genfromtxt('Generatornormaldot0_img/Generatornormaldot0_correler.txt', delimiter='')
my_datad2 = genfromtxt('Generatornormaldot0_img/Generatornormaldot0_dotter.txt', delimiter='')
my_datad3 = genfromtxt('Generatornormaldot0_img/Generatornormaldot0_dloss.txt', delimiter='')
my_datad4 = genfromtxt('Generatornormaldot0_img/Generatornormaldot0_gloss.txt', delimiter='')

my_datae1 = genfromtxt('Generatornormalext0_img/Generatornormalext0_correler.txt', delimiter='')
my_datae2 = genfromtxt('Generatornormalext0_img/Generatornormalext0_dotter.txt', delimiter='')
my_datae3 = genfromtxt('Generatornormalext0_img/Generatornormalext0_dloss.txt', delimiter='')
my_datae4 = genfromtxt('Generatornormalext0_img/Generatornormalext0_gloss.txt', delimiter='')
```

```python
def avgoverN(x,N):
    l = len(x)
    result = []
    for i in range (0,l,N):
        i_next = min(i + 100, l)
        if i_next == l:
            break
        result.append( np.mean(x[i:i_next]) )
   # print (result)
    return result
```

```python
plt.figure(figsize=(15,4))

plt.plot(avgoverN(my_data1,100),'b',label='base')
plt.plot(avgoverN(my_datal11,100),'r',label='L1')
plt.plot(avgoverN(my_datal21,100),'g',label='L2')
plt.plot(avgoverN(my_datae1,100),'b--',label='ext')
plt.plot(avgoverN(my_datac1,100),'r--',label='correl')
plt.plot(avgoverN(my_datad1,100),'g--',label='dot')
plt.legend()
plt.show()
```
