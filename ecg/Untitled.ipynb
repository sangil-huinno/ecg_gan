```python
def avgoverN(x,N,s):
    l = len(x)
    result = []
    for i in range (0,l,s):
        i_next = min(i + N, l)
        if i_next == l:
            break
        result.append( np.mean(x[i:i_next]) )
   # print (result)
    return result
```

```python
plt.figure(figsize=(15,4))

plt.plot(avgoverN(my_data1,300,50),'b',label='base')
plt.plot(avgoverN(my_datal11,300,50),'r',label='L1')
plt.plot(avgoverN(my_datal21,300,50),'g',label='L2')
plt.plot(avgoverN(my_datae1,300,50),'b--',label='ext')
plt.plot(avgoverN(my_datac1,300,50),'r--',label='correl')
plt.plot(avgoverN(my_datad1,300,50),'g--',label='dot')
plt.title('correlation between real and fake signals')
plt.legend()
plt.show()
```
